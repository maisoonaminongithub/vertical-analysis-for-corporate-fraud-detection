### Analysis and Visualization
We will be using two python libraries as well, maybe more if necessary. Ai based assistance and online converters to convert necessary files as assistance will also be taken.


```python
import pandas as pd

# Paths to the CSV files
file_path_A = '/kaggle/input/randomly-generated-company-income-statement/companyA.csv'
file_path_B = '/kaggle/input/randomly-generated-company-income-statement/companyB.csv'
file_path_C = '/kaggle/input/randomly-generated-company-income-statement/companyC.csv'

# Loading the CSV files into separate DataFrames
df_A = pd.read_csv(file_path_A)
df_B = pd.read_csv(file_path_B)
df_C = pd.read_csv(file_path_C)

# Printing the first few rows of each DataFrame to verify contents
print("DataFrame for Company A:")
print(df_A.head())
print("\nDataFrame for Company B:")
print(df_B.head())
print("\nDataFrame for Company C:")
print(df_C.head())
```

**We see here that the data is not clean. We will clean the data now. Remove commas and also convert all the numeric values to float.**

```python
# Function to convert columns to numeric, removing commas
def convert_to_numeric(df):
    for column in df.columns[1:]:  # Skip the 'Particulars' column
        df[column] = pd.to_numeric(df[column].replace('[\$,]', '', regex=True))
    return df

# Apply the conversion to each DataFrame
df_A_clean = convert_to_numeric(df_A)
df_B_clean = convert_to_numeric(df_B)
df_C_clean = convert_to_numeric(df_C)

df_A_clean.head(), df_B_clean.head(), df_C_clean.head()
```

**Here's how we'll perform vertical analysis for each of the three companies:** <br>
**Select the base figure:** In this case, we'll use 'TOTAL REVENUES' as the base figure for each year. <br>
**Calculate the percentage:** Each item on the income statement will be expressed as a percentage of the total revenues for the corresponding year.

Now lets construct the main function for verticle analysis

```python
# Function to perform vertical analysis
def vertical_analysis(df):
    for year in df.columns[1:]:  # Skip the 'Particulars' column
        df[year] = df[year] / df.loc[df['Particulars'] == 'TOTAL REVENUES', year].values[0] * 100
    return df

# Function to plot line graph for vertical analysis results
import matplotlib.pyplot as plt
def plot_line_graph(df, company_name):
    fig, ax = plt.subplots(figsize=(10, 6))
    years = df.columns[1:]  # Excluding the 'Particulars' column
    for index, row in df.iterrows():
        ax.plot(years, row[1:], label=row['Particulars'], marker='o')
    
    ax.set_ylabel('Percentage of Total Revenues (%)')
    ax.set_title(f'Vertical Analysis Trends for {company_name}')
    ax.legend(title='Income Statement Items')
    plt.grid(True)
```

```python
# Apply vertical analysis for Company A
df_A_vertical = vertical_analysis(df_A_clean.copy())
df_A_vertical.head()

# Plotting line graphs
plot_line_graph(df_A_vertical, "Company A")
plt.show()
```

```python
# Apply vertical analysis for Company B
df_B_vertical = vertical_analysis(df_B_clean.copy())
df_B_vertical.head()

# Plotting line graphs
plot_line_graph(df_B_vertical, "Company B")
plt.show()
```

```python
# Apply vertical analysis for Company C
df_C_vertical = vertical_analysis(df_C_clean.copy())
df_C_vertical.head()

# Plotting line graphs
plot_line_graph(df_C_vertical, "Company C")
plt.show()
```

# Summary
To summarize, vertical analysis is a method of examining financial statements in which each item in the important categories (balance sheet, income statement, etc.) is represented as a percentage of the overall amount within the same statement. This approach is commonly used to calculate the relative worth of numerous parameters and normalize comparisons over time or between businesses. For example, in an income statement, each item may be represented as a percentage of total revenue. This allows analysts to evaluate cost structures and profitability across firms of varying sizes. It is notably useful for producing trend lines for specific costs or revenues as a proportion of total sales across many time periods. Vertical analysis as a tool for detecting financial fraud might be part of a larger approach to financial statement review. While vertical analysis alone may not be adequate in identifying fraud, it may reveal unusual changes or abnormalities in financial statement items that warrant further investigation.
